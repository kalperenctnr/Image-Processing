```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt 
import helper
```

```python
path = r".\Segmentation\bird images\bird 1.jpg"
img = cv.imread(path)
img_gray = helper.convert2gray(img)
img = cv.GaussianBlur(img_gray, (3,3), 0)
```

```python
images = []
images.append(img)

k = 100
img_binary = helper.local_otsu(img, (k, k))
img_binary = 1- img_binary
images.append(img_binary.copy())

kernel = np.ones((7, 7))
img_binary = helper.erode(img_binary, kernel)
images.append(img_binary.copy())

kernel = np.ones((7, 7))
img_binary = helper.dilate(img_binary, kernel)
images.append(img_binary.copy())


# kernel = np.ones((7, 7))
# img_binary = helper.erode(img_binary, kernel)
# images.append(img_binary.copy())

output, labels = CCA(img_binary)
images.append(output)
print(f"Number of birds in the image : {len(labels)}")

titles = ["Gray Scale", "Thresholded", "Erosion", "Dilation", "CCA"]
fig, ax = plt.subplots(1, 5)

for i in range(5):
    ax[i].imshow(images[i], cmap='gray')
    ax[i].set_title(titles[i])
    
fig.set_figwidth(20)
fig.tight_layout()
plt.show()
```
